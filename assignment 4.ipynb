```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns 
import statsmodels.api as sm
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
x=df['sepal length (cm)']
y=df['petal length (cm)']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x,kde=True)
plt.show()
```

```python
sm.qqplot(x,line="s")
plt.show()
```

```python
stat,p=stats.shapiro(x)
print("shapiro:",stat)
print("p value:",p)
if p>0.05:
    print("dtata is normally distributed")
else:
    print("data is not normal")
```

```python
sns.boxplot(x)
plt.show()
```

```python
c=stats.skew(x)
print("symmetry by stat:",c)
```

```python
corr,pval=stats.spearmanr(x,y)
print("spearman :",corr)
pval value",pval)
```
